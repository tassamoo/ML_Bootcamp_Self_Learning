```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
3.29e7
```

```python
# Separate Data using DataFrame
X = df(data, columns = ['production_budget_usd'])
y = df(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,y, alpha = 0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3_000_000_000)
plt.xlim(0, 450_000_000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_ #theta_1
```

```python
#Intercept, theta_0
regression.intercept_
```

Plotted Regression:

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,y, alpha = 0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=4)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3_000_000_000)
plt.xlim(0, 450_000_000)
plt.show()
```

```python
#Rsquared - Check the variation it means the model can predict around 54% or 55%
regression.score(X, y)
```
